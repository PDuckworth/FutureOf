```python
# encoding=utf8
import os
import numpy as np
import pandas as pd
import getpass
import matplotlib.pyplot as plt
%matplotlib inline

datasets = '/home/'+ getpass.getuser() +'/Datasets/ONET/'
print datasets
d = os.path.join(datasets, 'FoHC/FOH Occupations Tasks Features Technology.xlsx')
excel_doc = pd.ExcelFile(d)
matts_data = excel_doc.parse("Title, Tasks, Features").fillna("-")

matts_data.rename(columns = {'Occupation title':'Observed Occupation'}, inplace = True)
matts_data.sort_values(by='Task')
```

```python
duplicate_task_rows = matts_data.copy()
for i, row in matts_data.iterrows():
    if row['Task also done by'] != '-': 
        row['Observed Occupation'] = row['Task also done by']
        duplicate_task_rows = duplicate_task_rows.append(row, ignore_index=True)

    if row['Task also done by.1'] != '-': 
        row['Observed Occupation'] = row['Task also done by.1']
        duplicate_task_rows = duplicate_task_rows.append(row, ignore_index=True) 
```

```python
duplicate_task_rows.sort_values(by = ['Observed Occupation', 'Task'], inplace=True)
duplicate_task_rows.drop(['Task also done by', 'Task also done by.1'], axis=1, inplace=True)
duplicate_task_rows.reset_index()
```

```python
# duplicate_task_rows.sort_values(by = ['Task', 'Observed Occupation'], inplace=True)
output_doc = os.path.join(datasets, 'FoHC/Duplicated_also_done_by_tasks.csv')
duplicate_task_rows.to_csv(output_doc, sep='\t', encoding='utf-8')
```
