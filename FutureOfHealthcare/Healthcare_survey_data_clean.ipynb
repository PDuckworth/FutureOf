```python
# encoding=utf8
import os
import numpy as np
import pandas as pd
import getpass
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',**{'family': 'serif', 'serif':"Computer Modern Roman"})
rc('text', usetex=True)

import seaborn as sns
from scipy.stats import gaussian_kde
%matplotlib inline
pd.options

## # Input DATA set of observed Occupations and Tasks
datasets = '/home/'+ getpass.getuser() +'/Datasets/'
out_data_path = "/home/scpd/Dropbox/Documents/Oxford/Programming/FutureOf/FutureOfHealthcare/"

d = os.path.join(datasets, 'FoHealthcare/250718-survey-workload-automation-desire-ratings.csv')

csv_doc = pd.read_csv(d)
data = csv_doc.iloc[2:]
```

```python
text_column_names = csv_doc.iloc[0,:]
```

```python
columns = csv_doc.columns
columns
```

```python
headers = []
non_missing = []
for i in range(len(columns)): 
    if not data.iloc[:,i].isnull().values.all():
        non_missing.append(data.iloc[:,i])
        headers.append(text_column_names[i])

df = pd.DataFrame(np.array(non_missing).T)

### add the headers
df.columns = headers
```

```python
out_data_path + 'healthcare_survey_clean.xlsx'
```

```python
writer = pd.ExcelWriter(out_data_path + 'Healthcare_survey_data/healthcare_survey_clean.xlsx', engine='xlsxwriter')
df.to_excel(writer)
writer.save()
```
